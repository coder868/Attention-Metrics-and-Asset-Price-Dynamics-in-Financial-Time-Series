```python
import statsmodels.api as sm
import pandas as pd
```

```python
trends = pd.read_csv('trends.csv',index_col='date')
stock_data = pd.read_csv('stock_trends.csv', index_col='Date')
chocolate_data = pd.read_csv('chocolate.csv',index_col='date')
data = pd.concat([trends,chocolate_data,stock_data],axis=1)
data=data.dropna()
data.index.name = 'date'
data.to_csv('data.csv')

attention = data[[
    'Hormuz',
    'oil price',
    'war',
    'Iran',
    'stock market',
    'Hormuz_diff',
    'oil price_diff',
    'war_diff',
    'Iran_diff',
    'stock market_diff' 
]]
```

```python
# model 1, predict market volatility by oil returns
y = data['vol_10_spy_t+1']
X = data['oil_ret']
X=(X-X.mean()) / X.std()

X= sm.add_constant(X)

model1 = sm.OLS(y,X).fit()
print(model1.summary())
```

```python
# plot market volatility vs oil returns (model 1)
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(12,6))


spy_vol_data = data['vol_10_spy_t+1']
spy_vol_data = (spy_vol_data-spy_vol_data.mean()) / spy_vol_data.std()
oil_ret_data = data['oil_ret']
oil_ret_data = (oil_ret_data-oil_ret_data.mean()) / oil_ret_data.std()
oil_ret_data = oil_ret_data.rolling(window=10).mean()

ax1.plot(data.index, spy_vol_data, label="SPY Volatility (10 day std)", alpha=0.8)
ax1.plot(data.index, oil_ret_data, label="Oil Returns (10 day avg)", alpha=0.8)

# ax2 = ax1.twinx()
# ax2.plot(data.index, df["war"], label="War Attention", color="red", alpha=0.3)

ax1.set_title("Volatility and Oil Returns Over Time")
ax1.legend()
plt.show()
```

```python
# model 2, predict market volatility with attention
y = data['vol_10_spy_t+1']
X = attention
X=(X-X.mean()) / X.std()
X = sm.add_constant(X)
model2 = sm.OLS(y,X).fit()
print(model2.summary())
```

```python
# plot market volatility vs raw attention (model 2)
keywords = [
    'Iran',
    'Hormuz',
    'war',
    'stock market',
    'oil price'
]

fig, axes = plt.subplots(len(keywords), 1, figsize=(12, 15), sharex=True)

spy_vol = data['vol_10_spy_t+1']
spy_vol = (spy_vol - spy_vol.mean()) / spy_vol.std()

for i, kw in enumerate(keywords):
    kw_data = data[kw]
    kw_data = (kw_data - kw_data.mean()) / kw_data.std()
    
    axes[i].plot(data.index, spy_vol, label="SPY Volatility", alpha=0.8)
    axes[i].plot(data.index, kw_data, label=kw, alpha=0.8)
    
    axes[i].set_title(f"Volatility vs {kw}")
    axes[i].legend()

plt.tight_layout()
plt.show()
```

```python
# plot market volatility vs attention diff (model 2)
keywords = [
    'Iran',
    'Hormuz',
    'war',
    'stock market',
    'oil price'
]
keywords = [i+'_diff' for i in keywords]

fig, axes = plt.subplots(len(keywords), 1, figsize=(12, 15), sharex=True)

spy_vol = data['vol_10_spy_t+1']
spy_vol = (spy_vol - spy_vol.mean()) / spy_vol.std()

for i, kw in enumerate(keywords):
    kw_data = data[kw]
    kw_data = (kw_data - kw_data.mean()) / kw_data.std()
    
    axes[i].plot(data.index, spy_vol, label="SPY Volatility", alpha=0.8)
    axes[i].plot(data.index, kw_data, label=kw, alpha=0.8)
    
    axes[i].set_title(f"Volatility vs {kw}")
    axes[i].legend()

plt.tight_layout()
plt.show()
```
